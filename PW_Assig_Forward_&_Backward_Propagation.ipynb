Q1. What is the purpose of forward propagation in a neural network?

Forward propagation is the process of passing input data through the layers of a neural network to obtain the output. The purpose is to compute the network's predictions based on the current weights and biases. This is essential for both training and inference, as it allows the network to map inputs to outputs.

Q2. How is forward propagation implemented mathematically in a single-layer feedforward neural network?

In a single-layer feedforward neural network, forward propagation can be mathematically represented as follows:

1. Input Layer: X is the input data matrix where each row is a sample and each column is a feature.
2. Weights and Biases: W is the weight matrix and b is the bias vector.
3. Linear Combination: Compute the linear combination of inputs and weights:

 - Z=XW+b
4. Activation Function: Apply an activation function f to the linear combination:
 - A=f(Z)

 Here, Z is the result of the weighted sum of inputs plus the bias, and
A is the output after applying the activation function.

Q6. What is the purpose of backward propagation in a neural network?

Backward propagation (backpropagation) is the process of calculating the gradient of the loss function with respect to each weight and bias in the network. This gradient is used to update the weights and biases during training to minimize the loss, thereby improving the network's performance. Backpropagation enables the network to learn from the errors made during forward propagation.

Q9. What are some common challenges or issues that can occur during backward propagation, and how can they be addressed?

Common challenges during backward propagation include:

1. Vanishing/Exploding Gradients: Gradients can become very small (vanish) or very large (explode), leading to slow or unstable training. This can be addressed by using techniques like batch normalization, gradient clipping, or changing the activation functions (e.g., using ReLU instead of sigmoid).
2. Overfitting: The model may perform well on training data but poorly on unseen data. Techniques like dropout, regularization (L1/L2), and early stopping can help mitigate overfitting.
3. Computational Efficiency: Backpropagation can be computationally intensive, especially for deep networks. Using optimized libraries, GPU acceleration, and mini-batch training can improve efficiency.
4. Learning Rate Issues: A poorly chosen learning rate can lead to slow convergence or divergence. Adaptive learning rate methods like Adam, RMSprop, or learning rate schedules can help address this issue.
