Q1. What is Flask Framework? What are the advantages of Flask Framework?

Ans:
Flask is a micro web framework for Python. It's designed to be lightweight and simple, making it an excellent choice for building web applications, particularly smaller to medium-sized projects. Here are some advantages of the Flask framework:
1. Simplicity: Flask is straightforward and easy to learn.
2. Flexibility: You can choose your components and extensions.
3. Extensibility: Add features as needed while keeping the app lightweight.
4. Active Community: Strong community support with numerous extensions available.
5. RESTful: Well-suited for building RESTful APIs.
6. Jinja2 Templating: Customizable and widely used template engine.
7. Werkzeug Integration: Built on the Werkzeug WSGI toolkit for low-level handling.
8. Scalability: Can be extended for larger and more complex applications.

Q3. What is App routing in Flask? Why do we use app routes?

Ans:
App routing in Flask refers to the process of mapping URL patterns to specific view functions in a Flask web application. It defines how different URLs (routes) are handled and which function should be executed when a user accesses a particular route. App routing is a fundamental concept in web development, and it's used to create different pages, actions, and endpoints within a web application.

We use app routes in Flask to:
1. Define how URLs map to view functions.
2. Organize and structure web application code.
3. Create different pages, actions, and endpoints.
4. Enable user interaction with various parts of the application.
5. Handle RESTful APIs for data retrieval and manipulation.

 Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

Ans:
In Flask, the 'url_for()' function is used for URL building. It generates a URL for a given view function and any arguments it needs based on the endpoint.



```python
from flask import Flask, url_for

app = Flask(__name)

@app.route('/')
def home():
    return 'Welcome to the homepage'

@app.route('/about')
def about():
    return 'About Us'

if __name__ == '__main__':
    with app.test_request_context():
        print(url_for('home'))  # Output: /
        print(url_for('about')) # Output: /about
```
